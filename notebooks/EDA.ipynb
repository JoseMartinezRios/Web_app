```python
df['DATE'] = pd.to_datetime(df['DATE'])
df['YEAR'] = df['DATE'].dt.year
df['month'] = df['DATE'].dt.month
df.info()
```

```python
# snake case for the dataframe columns
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

```python
# look for typos in the main dataframe
df['vehicle_make'].unique()
```

```python
# convert vehicle_make to lower and capital letter
df['vehicle_make'] = df['vehicle_make'].str.lower().str.strip().str.title()
df.head()
```

```python
# Change Location and vehicle_model to lower() and capital letter
df['location'] = df['location'].str.lower().str.strip().str.title()
df['vehicle_model'] = df['vehicle_model'].str.lower().str.strip().str.title()
df.head(20)
```

```python
# Replace nan values in column vehicle_model
df['vehicle_model'] = df['vehicle_model'].fillna("")
df.info()
```

```python
# Search typos in vehicle_model column
typos_model = df[~df['vehicle_model'].duplicated()]
typos_model
```

```python
# Store all typos in vehicle model to the vehicle maker, dictionary
typos_models_dict = {}
for index, rows in df.iterrows():
    current_year = rows["vehicle_model_year"]
    makers = rows["vehicle_make"]
    if current_year not in typos_models_dict:
        # initialize an emptry innerdictionary for this year
        typos_models_dict[current_year] = {}
    # filter the original dataframe (df) for non-null models of this maker and year
    known_models_df = df[
        (df["vehicle_model_year"] == current_year) &
        (df["vehicle_make"] == makers) &
        (df["vehicle_model"].notna())
        ]
        # get the unique models and convert to a list
    possible_models_list = known_models_df["vehicle_model"].unique().tolist()
    # add this maker and its list of possible models to the inner dictionary for the current year
    typos_models_dict[current_year][makers] = possible_models_list

typos_models_dict
```

```python
# Replace empty strings in the models list with a known model from the same maker and year
for year, makers in typos_models_dict.items():
    for maker, models in makers.items():
        # Find a replacement model (any non-empty one)
        replacement = next((m for m in models if m != ""), None)

        # Replace all empty strings with the replacement
        typos_models_dict[year][maker] = [
            replacement if m == "" else m for m in models]
typos_models_dict
```

Now That the main dataframe is empty of Nan values, typos, and other little issues, it's time to show some results based on the dataframe sellings.

First Let's see what are the best-sellers of each company. Than the best sellers by county. Than, the top 10 of the whole place (NOVA SCOTIA)

```python
# Best sellers by company
best_sellers = df.pivot_table(index='vehicle_make',
                               values='vehicle_model',
                                 aggfunc='count').reset_index()
best_sellers.sort_values(by='vehicle_model', ascending=False).head(20)
```

```python
# WHICH ONE ARE THE BEST SELLERS BY MODEL?
best_sellers_model = df.pivot_table(index='vehicle_model',
                                     values='vehicle_make',
                                       aggfunc='count').reset_index()
best_sellers_model.sort_values(by='vehicle_make', ascending=False).head(20)
```

```python
# County with most sales
best_sellers_county = df.pivot_table(index='location',
                                     values='vehicle_model',
                                       aggfunc='count').reset_index()
best_sellers_county.sort_values(by='vehicle_model', ascending=False).head(20)
```

```python
# Does the seasonality affect the sales?
seasonality = df.pivot_table(index='month',
                              values='vehicle_model',
                                aggfunc='count').reset_index()
seasonality
```

```python
# Year with most sales
yearly_sales = df.pivot_table(index='year',
                              values='vehicle_model',
                                aggfunc='count').reset_index()
yearly_sales
```
